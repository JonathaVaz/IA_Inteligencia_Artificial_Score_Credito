## Passo 2

```python
# Preparar a Base de Dados para a IA
# Importando o LabelEncoder e o Scikit-Learn

from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
tabela['Occupation'] = codificador.fit_transform(tabela['Occupation'])
tabela['Credit_Mix'] = codificador.fit_transform(tabela['Credit_Mix'])
tabela['Payment_Behaviour'] = codificador.fit_transform(tabela['Payment_Behaviour']) 
```

```python
tabela.info()
```

```python
tabela.dtypes
```

```python
tabela.columns
```

```python
# Separando as Informações em Base de Treino e Base de Teste

y = tabela['Credit_Score']
```

```python
# Todas as colunas que vamos usar para fazer a previsão do score credito
# Descartando Colunas Desnecessárias para Aprendizado do Modelo
# axis = 1 -> coluna
# axis = 0 -> linha 

x = tabela.drop(['Credit_Score', 'ID', 'Name', 'Num_Credit_Card', 'Month', 'Customer_ID', 'SSN', 'Age', 'Monthly_Inhand_Salary', 'Num_Credit_Inquiries', 'Total_EMI_per_month', 'Credit_Utilization_Ratio', 'Annual_Income', 'Num_of_Loan', 'Type_of_Loan', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Outstanding_Debt', 'Credit_History_Age', 'Payment_of_Min_Amount', 'Amount_invested_monthly', 'Monthly_Balance'], axis=1)
```

```python
# Separando as Informações pelo Sklearn Automaticamente "Padronizado"


from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

### Passo 2 - Ok
## Passo 3

```python
# Criando um Modelo de IA para Score de Créditos
# Arvore de Decisão 
# KNN

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvore = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

### Passo 3 - Ok
## Passo 4

```python
# Escolher o Melhor Modelo

from sklearn.metrics import accuracy_score

precisao_arvore = modelo_arvore.predict(x_teste)
precisao_knn = modelo_knn.predict(x_teste.to_numpy())

print(accuracy_score(y_teste, precisao_arvore))
print(accuracy_score(y_teste, precisao_knn))
```
